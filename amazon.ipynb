```python
df=pd.read_csv("amazon_sales.csv")
print(df)
```

```python
grouped_data=df.groupby('Category')['Total Sales'].sum()
result=pd.DataFrame({'Total Sales':grouped_data})
plt.plot(result, c="red",marker="o")
plt.xlabel("Category",c="blue",size=20)
plt.ylabel("Total sales",c="blue",size=20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,5))
sns.barplot(data=df,
           x="Payment Method",
           y="Quantity",
           hue="Payment Method",
           palette="deep")
plt.show()
```

```python
sns.countplot(x=df['Status'])
plt.xlabel("Order status")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create pivot table
pivot_table = df.pivot_table(index='Product', columns='Status', aggfunc='size',)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap of Online Order by Restaurant Type")
plt.xlabel("Status")
plt.ylabel("products")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("amazon_sales.csv")
print(df.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create pivot table
pivot_table = df.pivot_table(index='Product', columns='Status', aggfunc='size',)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap of Online Order by Status")
plt.xlabel("Status of products")
plt.ylabel("products")

plt.show()
```
