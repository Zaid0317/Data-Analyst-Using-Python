```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv('products.csv',encoding='ISO-8859-1')
print(df.head())
```

```python
sns.countplot(x=df['delivery_status'])
plt.xlabel("Types of Product")
sns.set_theme(font_scale=0.8)
```

```python
plt.hist(df['rating'],bins=5)
plt.title("Rating")
plt.show()
```

```python
# Create pivot table
pivot_table = df.pivot_table(index='category', columns='delivery_status', aggfunc='size',)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap of Category status")
plt.xlabel("status")
plt.ylabel("category")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('products.csv',encoding='ISO-8859-1')
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

grouped_data=df.groupby('category')['rating'].sum()
result=pd.DataFrame({'rating':grouped_data})
plt.figure(figsize=(14,5))
sns.violinplot(data=df,
              x="category",
              y="rating"
              )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

grouped_data = df.groupby('brand')['price'].sum()
result = pd.DataFrame({'price': grouped_data})

plt.figure(figsize=(15, 8))
plt.bar(result.index, result['price']) 
plt.xlabel('Brand')
plt.ylabel('Total Price')
plt.title('Total Price by Brand') 
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('products.csv',encoding='ISO-8859-1',low_memory=False)
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.figure(figsize=(18,5))
sns.boxplot(data=df,x='campaign_name',y='revenue_generated')
#plt.title("revenue generated")
plt.show()
```

```python
pivot_table = df.pivot_table(index='feedback_category', columns='rating', aggfunc='size',)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap of rating by feedback")
plt.xlabel("rating")
plt.ylabel("feedback_category")
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.lineplot(data=df,
           x="payment_method",
           y="revenue_generated",
           )
plt.show()
```
