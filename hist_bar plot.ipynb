```python
tips= sns.load_dataset('tips')
t=tips.groupby(["day", "sex"])["tip"].sum().reset_index()
t.head()
sns.set_theme(font_scale=1.2)

plt.figure(figsize=(14,5))
sns.barplot(data=tips, x="day", y="tip", hue="sex", order=["thur","fri","sat","sun"],palette="muted")

plt.title("Bar plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips= sns.load_dataset('tips')
sns.set_theme(font_scale=1.2)

plt.figure(figsize=(14,5))
sns.histplot(data=tips, x="total_bill", stat="density", kde="True", bins=25)

plt.title("histogram plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips= sns.load_dataset('tips')
sns.set_theme(font_scale=1.2)

plt.figure(figsize=(14,5))
sns.pairplot(data=tips, vars=["tip", "total_bill", "size"], kind="scatter", diag_kind="kde", hue="sex", height=2, aspect=1.1, palette="muted")

plt.suptitle("pair plot", y=1.02)
plt.show()
```
