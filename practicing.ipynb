```python
import pandas as pd

df=pd.read_csv("sample_data.csv")
print(df)

df.isnull().sum()
```

```python
#filling missing values for ML to work properly

#df[df['gender'].isnull()]
df[df.isnull().any(axis=1)]
#for passed column, fill NaN with mode
df['passed'].fillna(df['passed'].mode()[0], inplace=True)

#for gender column
df['gender'].fillna(df['gender'].mode()[0], inplace=True)

print(df)
```

```python
#converting categorical data(gender and passed columns) to numerical data
from sklearn.preprocessing import LabelEncoder

df1=df.copy()
le=LabelEncoder()

df1["gender_encoded"]=le.fit_transform(df1["gender"])
df1["passed_encoded"]=le.fit_transform(df1["passed"])

print("Encoded data is\n: ")
print(df1[['name', 'gender', 'gender_encoded', 'passed', 'passed_encoded']])
```

```python
#linear regression example
from sklearn.linear_model import LinearRegression

X= [[1], [2], [3], [4], [5]]
y= [[40], [50], [65], [75], [90]]

model= LinearRegression()

model.fit(X, y)

hours= eval(input("enter how many hours you studied: "))
predicted_marks=model.predict([[hours]])

print(f"based on your", hours, "hours you may score around", predicted_marks)

#three points to remember
# 1. its not about the line, its about the story 
# 2. 5 rows use start
# 3. accuracy is not always the goal, sometimes its about the story it tells
```

```python
# classification
#logistic regression
from sklearn.linear_model import LogisticRegression

X=[[1], [2], [3], [4], [5]] #hours studied ye h hamara input
y=[0,0,1,1,1] #result 0 aaya toh fail, 1 aaya to pass

model=LogisticRegression()

model.fit(X, y)

hours= eval(input("enter how many hours you studied: "))
result= model.predict([[hours]])[0]  # passing extra zero taki label nikal kr aaye 0 ya 1 

if result == 1:
    print(f"Based on your", hours, "hours studied You are likely to pass")

else:
    print(f"Based on your", hours, "hours studied You are likely to fail")
```

```python
#K nearest neighbors

from sklearn.neighbors import KNeighborsClassifier

X=[[180,7], #fruit weight in grams, size in cm 180 grams, and 7 cm size
  [200, 7.5],
  [250, 8],
  [300, 8.5],
  [330, 9],
  [360, 9.5]]

#0= apple, 1=orange
y=[0,0,0,1,1,1]

model= KNeighborsClassifier(n_neighbors=3)

model.fit(X, y)
weight=eval(input("enter the weight in grams"))
size=eval(input("enter the size in centimeters"))

prediction=model.predict([[weight, size]])[0]

if prediction == 0:
    print("this is likely an Apple")

else:
    print("this is likely an orange")
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

X=[[7, 2], #apple
  [8, 3], #apple
  [9, 8], #orange
  [10, 9]] #orange

y=[0,0,1,1] #0= apple, 1=orange
model= DecisionTreeClassifier()

model.fit(X, y)

#color_shade= 1-3= red, 4-10=orange color
size=eval(input("enter the fruits size in cm"))
color_shade=eval(input("enter the color shade (1-10): "))

result=model.predict([[size,color_shade]])[0]

if result == 0:
    print("this is likely an Apple")

else:
    ("this is likely an orange")

'''
three secret:

1. machine ko jitna ache se aap clear rule doge unta acha se wo kaam kregi
2. trees can be too smart, mtlb jitna zyda aap training data doge, tree jo hai utna zyda khud se memorize krte jyegi
3. it is easy to explain to boss or colleague
'''
```

```python
#mastering model evaluation
#classification Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#true answers(what actually happened)
y_true=[1,0,1,1,0,1,0]
#model prediction(what it guessed)
y_pred=[1,0,1,0,0,1,1]

#evaluation
print("Accuracy: ", accuracy_score(y_true, y_pred))

print("Precision: ", precision_score(y_true, y_pred)) #agr maine 10 mei se 8 logon ka prediction kiya tha, toh unn 8 mei se kitne prediction 
                                                        #acutally mei sahi the

print("Recall: ", recall_score(y_true, y_pred)) # actual mei kitne real positives mujhe mile

print("F1 score: ", f1_score(y_true, y_pred)) # agr mai precision and recall ko mix krdun toh kya number niklega

'''
three points to remember:
1. accuracy is not enough if data is unbalanced
2.precision ka use kriye when False alarm are dangerous
3. aap recall ka use kriye when missing case is dangerous'''
```

```python
 #regression metrics: MAE, MSE, RMSE
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

#real score from test result
real_score= [90,60,80,100]

#model guess
predicted_score=[85,70,70,95]

mae=mean_absolute_error(real_score, predicted_score)
mse=mean_squared_error(real_score, predicted_score)
rmse=np.sqrt(mse)

print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

data=pd.read_csv('student.csv')

X=data[['hours']] #double brackets= 2d input
y=data['score'] #target column

model= LinearRegression()
model.fit(X, y)

predicted_scores=model.predict(X)

#evaluate
mae=mean_absolute_error(y, predicted_scores)
mse=mean_squared_error(y, predicted_scores)
rmse=np.sqrt(mse)

#result
print("Mean Absolute Error (MAE): ", mae)
print("Mean Squared Error (MSE): ", mse)
print("Root Mean Squared Error (RMSE): ", rmse)

new_hour=eval(input("enter the number of hours you studied: "))
new_pred= model.predict([[new_hour]])
print(f"prediction for ", new_hour, "hours is: ", new_pred)
```
