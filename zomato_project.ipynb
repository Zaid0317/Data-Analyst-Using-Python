```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("project.csv",encoding='ISO-8859-1')
print(df.head())
```

```python
df
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("types of restaurant")
```

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="yellow",marker="o")
plt.xlabel("type of restaurant",c="blue",size=20)
plt.ylabel("Votes",c="blue",size=20)
```

```python
plt.figure(figsize(14,5))
plt.hist(df['rate'],bins=5)
plt.title("rating distribution")
plt.show()
```

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df,x='online_order',y='rate')
```

```python
import pandas as pd

df=pd.read_csv("project.csv",encoding='ISO-8859-1')
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create pivot table
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size',)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap of Online Order by Restaurant Type")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
